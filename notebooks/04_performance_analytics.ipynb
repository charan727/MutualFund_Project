```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
```

```python
nav_df = pd.read_csv("../data/processed/clean_nav_history.csv")
benchmark_df = pd.read_csv("../data/raw/benchmark_indices.csv")

print(nav_df.shape)
print(benchmark_df.shape)
```

```python
nav_df['date'] = pd.to_datetime(nav_df['date'])

nav_df = nav_df.sort_values(
    ['amfi_code', 'date']
)

nav_df['daily_return'] = (
    nav_df.groupby('amfi_code')['nav']
          .pct_change()
)

nav_df.head()
```

```python
rf = 0.065

sharpe_df = (
    nav_df.groupby('amfi_code')['daily_return']
    .agg(['mean','std'])
    .reset_index()
)

sharpe_df['sharpe_ratio'] = (
    ((sharpe_df['mean'] * 252) - rf)
    /
    (sharpe_df['std'] * np.sqrt(252))
)

sharpe_df
```
